# Numpy

```python
import numpy as np

array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array created with np.array:\n", array1)
array2 = np.zeros((2, 3))
array2[0, :] = [1, 2, 3]
array2[1, :] = [4, 5, 6]
print("Array created with np.zeros and manual assignment:\n", array2)
array3 = np.arange(1, 7).reshape((2, 3))
print("Array created with np.arange and np.reshape:\n", array3)
```

```python
array_1d = np.linspace(1, 10, 100)
array_2d = array_1d.reshape((10, 10))

print("1D array of 100 evenly spaced numbers between 1 and 10:\n", array_1d)
print("\nReshaped 2D array (10x10):\n", array_2d)
```

#### np.array always creates a new array and copies the data.
#### np.asarray converts input to an ndarray without copying if the input is already an ndarray.
#### np.asanyarray is like np.asarray but preserves subclasses.
#### Deep copy creates an entirely independent copy of an object, whereas shallow copy creates a new object but with references to the original object's elements.

```python
import numpy as np
a = [1, 2, 3]
b = np.array(a)
print(b) 
```

```python
a = np.array([1, 2, 3])
b = np.asarray(a)
print(b is a) 
```

```python
a = np.matrix([1, 2, 3])
b = np.asanyarray(a)
print(type(b))  
```

```python
import copy
a = [1, 2, [3, 4]]
b = copy.deepcopy(a)
b[2][0] = 10
print(a)  
print(b) 
```

```python
import copy
a = [1, 2, [3, 4]]
b = copy.copy(a)
b[2][0] = 10
print(a) 
print(b) 
```

# visualization

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, color='red', marker='o')
plt.axhline(y=0.5, linestyle='--', color='blue', label='y = 0.5')
plt.axvline(x=0.5, linestyle=':', color='green', label='x = 0.5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')
plt.legend()
plt.show()
```

```python
np.random.seed(0)
dates = pd.date_range(start='2024-01-01', periods=100)
temperature = np.random.uniform(10, 30, size=100)
humidity = np.random.uniform(40, 80, size=100)
plt.figure(figsize=(10, 6))

# Plot temperature on left y-axis
plt.plot(dates, temperature, color='red', label='Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)', color='red')

# Create a second y-axis for humidity
plt.twinx()
plt.plot(dates, humidity, color='blue', label='Humidity')
plt.ylabel('Humidity (%)', color='blue')
plt.title('Temperature and Humidity Over Time')
plt.legend()
plt.show()
```

```python
from scipy.stats import norm
data = np.random.normal(size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
import seaborn as sns
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

quadrant = np.zeros_like(x, dtype=int)
quadrant[(x >= 0) & (y >= 0)] = 1  # Quadrant I
quadrant[(x < 0) & (y >= 0)] = 2   # Quadrant II
quadrant[(x < 0) & (y < 0)] = 3    # Quadrant III
quadrant[(x >= 0) & (y < 0)] = 4   # Quadrant IV

colors = ['red', 'blue', 'green', 'orange']
color_map = {1: colors[0], 2: colors[1], 3: colors[2], 4: colors[3]}
point_colors = [color_map[q] for q in quadrant]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=quadrant, palette=colors, legend='full')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
```

```python
from bokeh.plotting import figure, show
import numpy as np
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5
p.line(x, y, line_width=2)
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(1, 20, size=len(categories))
p = figure(x_range=categories, height=350, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values')

colors = ['#FF5733', '#33FF57', '#3366FF', '#FF33E6']
source = ColumnDataSource(data=dict(categories=categories, values=values, colors=colors))
p.vbar(x='categories', top='values', width=0.5, color='colors', legend_label='values', source=source)

tooltips = [('Category', '@categories'), ('Value', '@values')]
p.add_tools(HoverTool(tooltips=tooltips))

p.xaxis.major_label_orientation = np.pi / 4  # Rotate x-axis labels
p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'
show(p)
```
